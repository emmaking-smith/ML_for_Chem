```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

def root_mean_squared_error(true, pred):
    mean_squred_error = np.mean((true - pred)**2)
    return np.sqrt(mean_squred_error)
```

```python
df = pd.read_csv('doyle.csv')
df
```

```python
rxn_fps = np.load('rxn_fps_inputs.npy')
```

```python
labels = np.array(df['yield'])
```

```python
rf = RandomForestRegressor()
```

```python
rxn_fps = StandardScaler().fit_transform(rxn_fps)
```

```python
train_inputs, test_inputs, train_labels, test_labels = train_test_split(
rxn_fps,
    labels,
    test_size=0.33,
    random_state=12
)
```

```python
rf.fit(train_inputs, train_labels)
```

```python
preds = rf.predict(test_inputs)
```

```python
mean_absolute_error(test_labels, preds)
```

```python
root_mean_squared_error(test_labels, preds)
```

```python
r2_score(test_labels, preds)
```
